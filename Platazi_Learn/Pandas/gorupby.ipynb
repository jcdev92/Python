## GroupBy

```python
import pandas as pd
```

```python
df_books = pd.read_csv('bestsellers-with-categories.csv')
df_books.head(2)
```

```python
df_books.groupby('Author').count() # agregamientos funciones de agregacion + count / contar
```

```python
df_books.groupby('Author').mean() # the mean - la media
```

```python
df_books.groupby('Author').min()
```

```python
df_books.groupby('Author').max()
```

```python
df_books.groupby('Author').sum()
```

```python
df_books.groupby('Author').sum().loc['William Davis']
```

```python
df_books.groupby('Author').sum().reset_index()
```

```python
df_books.groupby('Author').agg(['min','max'])
```

```python
df_books.groupby('Author').agg({'Reviews':['min','max'],'User Rating':'sum'})
```

```python
df_books.groupby(['Author','Name']).count()
```

```python
df_books.groupby(['Author','Name']).agg({'Reviews':['min','max'],'User Rating':'sum'})
```

```python
df_books.groupby(['Author','Name']).agg({'Reviews':['min','max'],'User Rating':'sum'}).reset_index()
```

```python
df_books.groupby(['Author','Name']).count().reset_index().sort_values('Reviews',ascending=False)
```
