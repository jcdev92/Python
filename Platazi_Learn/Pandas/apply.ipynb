```python
import pandas as pd
import numpy as np
df_books = pd.read_csv('./bestsellers-with-categories.csv')
```

### APPLY

```python
def two_times(value):
    return value*2
```

```python
df_books['User Rating'].apply(two_times)
```

```python
df_books['Double_Rating']= df_books['User Rating'].apply(two_times)
```

```python
df_books
```

```python
df_books_2 = df_books.copy()
```

```python
df_books_2
```

```python
df_books_2['Triple_Rating'] = df_books['User Rating'].apply(lambda x: x*3) # USING APPLY WHIT LAMBDA FUNCTION 
df_books_2
```

```python
df_books_2['Triple_Rating'] = df_books.apply(lambda x : x['User Rating']*3 if x['Genre']== 'Fiction' else x['User Rating'], axis=1) ## another more advanced option
df_books_2
```
