```python
links_list = []
sec = requests.get(directions[1])
```

```python
def make_lists(sec):
    s_section = BeautifulSoup(sec.text, 'lxml')
    groupes = s_section.find_all('div', attrs={'class' : 'article-item__content'})
    groupes
    for i in groupes:
        links_list.append(i.a.get('href'))
```

```python
make_lists(sec)
for i in range(len(links_list)):
    links_list[i] = f'https://www.pagina12.com.ar{links_list[i]}'
links_list

    
```

```python
import os
import datetime
```

```python
def notes_extartor(url_link):
    for i in links_list:
        try:
            notice = requests.get(i)
            if notice.status_code == 200:
                s_notice = BeautifulSoup(notice.text, 'lxml')
                
                # title path
                tp = s_notice.find('div',{'class':'col 2-col'})
                title = tp.find('h1')
                title = title.get_text()
                
                # paragraph path
                pp = s_notice.find('div',{'class':'article-text'})
                paragraph = [i.get_text() for i in pp]
                paragraph = '\n'.join(paragraph)
                
                # date path
                dp = s_notice.find('div',{'class':'date modification-date'})
                date = dp.span.get_text()
                
                # article prefix path
                att = s_notice.find('div', attrs = {'class':'col 2-col'})
                feed = att.h4.get_text()
                
                # author Path
                ap = tp.find('div', attrs = {'class':'author'})
                author = ap.get_text()
                
                # image extractions Path
                root = s_notice.find('div',{'class':'section-2-col article-main-image'})
                root = root.find('div',{'class':'col 2-col'})
                try: 
                    root = root.find('img',{'class':''})
                    image_src = root.get('src')
                except Exception as e:
                    print(f'{e}')
                    print(f'bad link: {url_link}')
                
                # saving news
                economics_news = datetime.date.today().strftime('%d-%m-%Y')
                if not os.path.isdir(economics_news):
                    os.mkdir(economics_news)
                with open(f'{economics_news}/{title}.txt','w',encoding='utf-8') as f:
                    f.write(f'{date} \n\n {feed} \n\n {title} \n\n {image_src} \n\n {paragraph} \n\n {author}')
            else:
                raise ValueError(f'Error: {notice.status_code} website failed: {url_link}')
    
        except Exception as e:
            print(f'error: {e}')
        
```

```python
notes_extartor(links_list)
```
